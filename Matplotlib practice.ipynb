```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [12,34,54,56,67,89]
x1 =[10,25,35,66,67,134]
y = [9,34,98,76,84,209]
a  = np.random.randint(1,100,10)
a1  = np.random.randint(1,100,10)
b  = np.random.randint(1,70,10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,color = "m", marker = "*", linestyle = "--")
axes.set_title("Table")
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,color = "m", marker = "*", linestyle = "--")
axes.plot(x1,y,color = "c", marker = "*", linestyle = "--")
axes.set_title("Table")
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_title("Graph")
axes.plot(a,b,color = "c", marker = "p", ls = "--", lw = 1, label = "x lable")
axes.plot(a1,b,marker ="*", ls = "--", lw = "1", color = "r", label = "y lable")
axes.legend()
```

```python
p = np.random.randint(10,100,10)
q = np.random.randint(10,100,10)
r = np.random.randint(10,100,10)
s = np.random.randint(10,100,10)
t = np.random.randint(10,100,10)
u = np.random.randint(10,100,10)
p.sort(),q.sort(),r.sort(),s.sort(),t.sort(),u.sort()
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].plot(p,q, label = "x",marker = "*")
axes[0].plot(u,q, label = "y", marker = "*")
axes[1].plot(r,s,label = "x",marker = "*")
axes[1].plot(r,p, label = "y", marker = "*")
axes[2].plot(t,u,label = "x",marker = "*")
axes[2].plot(q,u, label = "y", marker = "*")
plt.show()
```

```python
p = np.random.randint(10,100,10)
q = np.random.randint(10,100,10)
r = np.random.randint(10,100,10)
s = np.random.randint(10,100,10)
t = np.random.randint(10,100,10)
u = np.random.randint(10,100,10)
p.sort(),q.sort(),r.sort(),s.sort(),t.sort(),u.sort()
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].hist(q)
axes[0].set_title("Histogram")
axes[1].pie(s)
axes[1].set_title("Pie chart")
axes[2].scatter(t,p)
axes[2].set_title("Scatter plot")
```
